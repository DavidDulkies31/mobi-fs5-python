```python
import numpy as np
import matplotlib.pyplot as plt
```

**[Task 1.2 (a).]()** Use `plt.imread` to load an image:

```python
 img = plt.imread("data/NIH3T3/im/dna-0.png")
```

**[Task 1.2 (b).]()** Show the image:

```python
plt.figure(figsize=(15,8))
plt.imshow(img, cmap='gray')
plt.colorbar()
```

**[Task 1.2 (d).]()** Binarize the image by applying an intensity threshold:

```python
threshold = 0.1
binary_img = img > threshold
plt.imshow(binary_img, cmap='gray')
```

**[Task 1.2 (e).]()** Load the ground truth:

```python
ground_truth = plt.imread("data/NIH3T3/gt/0.png")
plt.imshow(ground_truth, cmap='gray')
```

Compute and print the Dice coefficient:

```python
dice_score = 2 * np.sum(binary_img * ground_truth) / (np.sum(binary_img) + np.sum(ground_truth))
print("Dice Score:", dice_score)
```
