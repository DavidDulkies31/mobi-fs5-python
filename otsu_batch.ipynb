**[Task 2.1.]()** Implement Otsu thresholding as a re-usable function:

```python
#definition of otsu thresholding
def otsu_thresholding(imgf):
    img8 = (imgf * 255).round().astype(np.uint8)

    h = []
    for i in range(0, 256) :
        count = np.sum(img8 == i)
        h.append(count)

    threshold = 0
    sigma_b_squared_max = 0


    for T in range(1, 255) :
        n1 = np.sum(h[0:T])       
        n2 = np.sum(h[T:256])
        if n1 == 0 or n2 == 0 :
            continue
        mu1 = np.sum(h[0:T] * np.arange(T)) / n1 
        mu2 = np.sum(h[T:256] * np.arange(T, 256)) / n2 
        sigma_b_squared = n1 * n2 * (mu1 - mu2) ** 2
   
        if sigma_b_squared > sigma_b_squared_max :
            sigma_b_squared_max = sigma_b_squared
            threshold = T
    
    binary_img = (img8 >= threshold)
    seg = binary_img.astype(np.uint8) * 255

    print(threshold)
    return seg

```

**[Task 2.2.]()** Implement a re-usable function to compute the Dice coefficient:

```python
def compute_dice(bin1, bin2):
    assert bin1.dtype == bool
    assert bin2.dtype == bool
    
    dice_score = 2 * np.sum(bin1 * bin2) / (np.sum(bin1) + np.sum(bin2))    
    return dice_score
```

**[Task 2.4.]()** Perform batch processing:

```python
image_indices = [28, 29, 33, 44, 46, 49]

dice_scores = []

for i in image_indices:
    img = plt.imread(f'data/NIH3T3/im/dna-{i}.png')
    ground_truth = plt.imread(f'data/NIH3T3/gt/{i}.png')

    seg = otsu_thresholding(img)
    binary_img = seg.astype(bool)

    dice = compute_dice(binary_img, ground_truth.astype(bool))
    dice_scores.append(dice)
    print(f"Dice coefficient for image {i}: {dice}")


mean_dice = np.mean(dice_scores)
print(f"\nMean Dice coefficient for all images: {mean_dice}")
```
