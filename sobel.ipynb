**[Task 2.2.]()** Implement the two functions `sobel_h` and `sobel_v` for horizontal and vertical $3 \times 3$ Sobel filtering, respectively:

```python
def sobel_h(img):
    Kx = np.array([[ -1, 0, 1],
                   [ -2, 0, 2],
                   [ -1, 0, 1]])
    rows, cols = img.shape
    img_h = np.zeros((rows, cols))
    for i in range(1, rows-1):
        for j in range(1, cols-1):
            region = img[i-1:i+2, j-1:j+2]
            img_h[i, j] = np.sum(Kx * region)
    return img_h
    
    
def sobel_v(img):
    Ky = np.array([[ 1, 2, 1],
                   [ 0, 0, 0],
                   [ -1, -2, -1]])
    rows, cols = img.shape
    img_v = np.zeros((rows, cols))
    for i in range(1, rows-1):
        for j in range(1, cols-1):
            region = img[i-1:i+2, j-1:j+2]
            img_v[i, j] = np.sum(Ky * region)
    return img_v
    
```

**[Task 2.3.]()** Test your implementations from above:

```python
# test implementations
img_h = sobel_h(image)
img_v = sobel_v(image)
plt.imshow(img_h, cmap='gray')
plt.imshow(img_v, cmap='gray')
```

**[Task 2.4.]()** Compute the image gradient magnitude:
